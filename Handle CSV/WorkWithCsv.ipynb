#14 Convertors

```python
def rename(name):
  if (name == "Royal Challengers Bangalore"):
      return "RCB"
  else:
      return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('/content/sample_dataset_250.csv' , converters={'team1' : rename} )
```
