```python
import requests
import pandas as pd

url = "https://movie-database-api1.p.rapidapi.com/list_movies.json?limit=20"

headers = {
    "x-rapidapi-key": "YOUR_API_KEY",
    "x-rapidapi-host": "movie-database-api1.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

data = response.json()

print(data)

# Optional
df = pd.json_normalize(data)
print(df.head())
```
