## Working with Dates

```python
# converting to datetime datetype

date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

#### 1 . Extract year

```python
date['date_year'] = date['date'].dt.year
date.head()
```

#### 2 . Extract Month

```python
date['date_month_no'] = date['date'].dt.month
date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

#### 3 . Extract Days

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
# Day of week 

date['date_dow'] = date['date'].dt.dayofweek
date.head()
```

```python
date['date_dow_name'] = date['date'].dt.day_name()
date.head()
```

```python
# is weekend ? 

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday' , 'Satruday']), 1,0)
date.drop(columns=['product_id' , 'city_id' , 'orders']).head()
```

#### 5 . Extract week of the year

```python
date['date_week'] = date['date'].dt.isocalendar().week
date.drop(columns=['product_id' , 'city_id' , 'orders']).head()
```

```python
# Extract Quarter 
date['quarter'] = date['date'].dt.quarter
date.drop(columns=['product_id' , 'city_id' , 'orders']).head()
```

```python
# Extract Semester
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)
date.drop(columns=['product_id' , 'city_id' , 'orders']).head()
```

#### 6 . Extract Time elapsed between dates

```python
import datetime
today = datetime.datetime.today()
today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
today = pd.Timestamp.today()

date['months_passed'] = (
    (today.year - date['date'].dt.year) * 12
    + (today.month - date['date'].dt.month)
)
date.head()
```
